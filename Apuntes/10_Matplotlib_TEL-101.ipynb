- Matplotlib es una biblioteca para crear visualizaciones en Python.
- Permite crear gráficos de calidad científica con solo unas pocas líneas de código.

### Primeros pasos
- Descargar e instalar la biblioteca [Matplotlib](https://matplotlib.org/).
- Importar el módulo `matplotlib` en su código Python.

```python
import matplotlib
```

- En el módulo `matplotlib`, existe un conjunto de funciones llamado `pyplot`.
- Para realizar cualquier acción gráfica se debe importar `pyplot` desde el módulo `matplotlib`.
- Luego, llamar a las funciones dentro de `pyplot`. Por ejemplo `pyplot.plot`.

```python
from matplotlib import pyplot
```

## Crear un gráfico
La función `pyplot.plot(x,y)` se utiliza para formar una figura con las coordenadas $x$ e $y$ de puntos (marcadores). Por defecto, la función no dibuja los puntos, sino que muestra una línea azul que conecta las coordenadas entregadas.

Parámetros:
- `x`: (Opcional) es una secuencia con las coordenas del **eje-x**. Por defecto son $0,1,2,3,\dots$
- `y`: (Obligatorio) es una secuencia con las coordenas del **eje-y**.

## Mostrar la figura

La función `pyplot.show()` se utiliza para mostrar la figura, generalmente en una nueva ventana. Si además desea guardar la figura en un archivo de imagen, use `pyplot.savefig(nombre)` antes de `pyplot.show()`.

## Gráfico de líneas

```python
x = [4, 8, 13, 17, 20]
y = [54, 67, 98, 78, 45]

pyplot.plot(x,y)
pyplot.savefig("grafico.pdf")
pyplot.show()
```

## Múltiples curvas en un gráfico

```python
y1 = [3, 8, 1, 10]
y2 = [6, 2, 7, 11]

pyplot.plot(y1, label="Serie 1")
pyplot.plot(y2, label="Serie 2")

pyplot.legend()
pyplot.show()
```

## Marcadores

La función `plot()`, además puede recibir un texto en el formato `Marcador|Línea|Color` que define el estilo visual de los puntos.

Todos los estilos disponibles para un marcador los puede encontrar en la siguiente Tabla: https://matplotlib.org/stable/api/markers_api.html

```python
y = [3, 8, 1, 10]
pyplot.plot(y, 'o:r')
pyplot.show()
```

El tamaño del marcador puede controlarse con el parámetro `ms`, mientras que el color del borde con `mec` y el color de relleno con `mfc`.

```python
y = [3, 8, 1, 10]

pyplot.plot(y, 'o-', ms=20, mec='r', mfc='g')
pyplot.show()
```

En tanto, el grosor de la línea puede controlarse con el parámetro `lw`.

```python
y = [3, 8, 1, 10]

pyplot.plot(y, lw = 20)
pyplot.show()
```

## Títulos y cuadrícula

Puede usar las funciones `xlabel()` e `ylabel()` para establecer una etiqueta para cada eje y `title()` para agregar un título al gráfico. 

Además, puede usar la función `grid()` para incluir líneas de cuadrícula en el gráfico.

```python
x = [4, 8, 13, 17, 20]
y = [54, 67, 98, 78, 45]
pyplot.figure(dpi=100)
pyplot.plot(x,y,'g--*',lw=1,ms=15)
pyplot.title("Título")
pyplot.xlabel("Eje x")
pyplot.ylabel("Eje y")
pyplot.grid(color = 'gray', linestyle = '--', linewidth = 0.2)
pyplot.show()
```

## Gráfico de dispersión

Se utiliza para mostrar coordenadas con valores continúos en dos dimensiones $(x,y)$.

```python
x = [2,4,6,7,9,13,19,26,29,31,36,40,48,51,57,67,69,71,78,88]
y = [54,72,43,2,8,98,109,5,35,28,48,83,94,84,73,11,464,75,200,54]
```

```python
pyplot.scatter(x,y)
pyplot.show()
```

## Gráfico de burbuja

Es un gráfico de dispersión en el cual es posible representar más de dos dimensiones. El tamaño del punto indica el valor de la tercera coordenada de los datos.

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
z = [20,50,100,200,500,1000,60,90,10,300,600,800,75]

pyplot.scatter(x, y, s=z, alpha=0.5)
pyplot.show()
```

## Burbujas de colores
El color del punto permite representar una cuarta coordenada en el gráfico de burbujas.

```python
from random import sample
x = sample(range(0,500),50)
y = sample(range(0,500),50)
z = sample(range(0,500),50)
colores = sample(range(0,100),50)
pyplot.scatter(x, y, s=z, c=colores, alpha=0.5, cmap='nipy_spectral')
pyplot.colorbar()
pyplot.show()
```

## Gráfico de barras
Un gráfico de barras o columnas es una forma de resumir un conjunto de datos por categorías.

```python
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

pyplot.bar(x,y)
pyplot.show()
```

## Gráfico circular
Muestra los datos en sectores para conocer su proporción con respecto al total.

```python
y = [15, 40, 60, 20]
etiquetas = ["Tortugas","Hamsters","Perros","Gatos"]

pyplot.figure(dpi=100)
pyplot.title("Mascotas")
pyplot.pie(y, labels=etiquetas, autopct="%.2f%%")
pyplot.show() 
```

## Subgráficos
La función `subplot()` recibe tres argumentos que describen la cantidad de filas, columnas e índice del gráfico actual.

```python
#Subgráfico 1:
pyplot.subplot(1, 2, 1) # 1 Fila, 2 Columnas, Subgráfico 1
pyplot.plot([3, 8, 1, 10]) # Puntos del Subgráfico 1
#Subgráfico 2:
pyplot.subplot(1, 2, 2) # 1 Fila, 2 Columnas, Subgráfico 2
pyplot.plot([10, 20, 30, 40], 'r-o') # Puntos del Subgráfico 2
pyplot.show()
```
